```python
# Function to read the file and analyze the lines
def analyze_file(file_path):
    with open(file_path, 'r') as file:
        lines = file.readlines()

    # Initialize counters
    total_lines = len(lines)
    separator_counts = []

    # Analyze each line
    for line in lines:
        # Strip any extra spaces or newline characters
        clean_line = line.strip()
        
        # Count the number of separated strings in the line
        separated_strings = clean_line.split('.')
        num_separated = len(separated_strings)
        
        # Store the result for this line
        separator_counts.append((clean_line, num_separated))

    # Return the total number of lines and the counts for each line
    return total_lines, separator_counts

# Specify the path to your text file
file_path = 'input.txt'

# Call the function
total_lines, separator_counts = analyze_file(file_path)

# Output the results
print(f"Total lines in file: {total_lines}")
for i, (line, count) in enumerate(separator_counts):
    print(f"Line {i + 1}: '{line}' has {count} separated strings.")
```

```python
def check_line_format_consistency(text_lines):
    inconsistent_lines = []
    
    for i, line in enumerate(text_lines, start=1):
        line = line.strip()
        
        # Skip empty lines
        if not line:
            continue

        # Split the line into parts
        parts = line.split()
        
        # Check if the line has exactly two parts
        if len(parts) != 2:
            inconsistent_lines.append((i, line, "Incorrect number of parts"))
            continue
        
        part1, part2 = parts
        
        # Check if each part contains a dot
        if '.' not in part1 or '.' not in part2:
            inconsistent_lines.append((i, line, "Missing dot in one or both parts"))
            continue
        
        # Check if each part has characters before and after the dot
        part1_split = part1.split('.')
        part2_split = part2.split('.')
        
        if any(not segment for segment in part1_split) or any(not segment for segment in part2_split):
            inconsistent_lines.append((i, line, "Empty segment before or after dot"))
    
    return inconsistent_lines
```
